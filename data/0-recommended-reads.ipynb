```python
df_cleaned = df.dropna(subset=["Book_Description", "Book_Title"])
df_cleaned
```

```python
# SÃ¼tun isimlerini kÃ¼Ã§Ã¼k harfli hale getirin
df_cleaned.columns = map(str.lower, df_cleaned.columns)

# Aradaki boÅŸluklarÄ± veya "_" karakterlerini kaldÄ±rÄ±n
df_cleaned.columns = df_cleaned.columns.str.replace(' ', '').str.replace('_', '')
```

```python
df_cleaned
```

```python
df_cleaned['untokenizeddata'] = df_cleaned['booktitle'] + ' ' + df_cleaned['authorname'] + ' ' + df_cleaned['bookdescription'] + ' ' + df_cleaned['genres'] + ' ' + df_cleaned['type']
```

```python
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

def clean_text(text):
    text = text.lower() # tÃ¼m karakterleri kÃ¼Ã§Ã¼k harf yap
    text = re.sub(r'\d+', '', text) # sayÄ±larÄ± Ã§Ä±kar
    text = ''.join(ch for ch in text if ch not in punctuation) # noktalama iÅŸaretlerini Ã§Ä±kar
    tokens = word_tokenize(text) # kelime Ã¶beklerine ayÄ±r
    tokens = [w for w in tokens if not w in stop_words] # stop-words'leri Ã§Ä±kar
    return tokens
```

```python
df_cleaned['tokenizeddata'] = df_cleaned['untokenizeddata'].apply(lambda x: clean_text(x))
```

```python
df_cleaned.to_csv("tokenizeddata.csv", index=False)
```

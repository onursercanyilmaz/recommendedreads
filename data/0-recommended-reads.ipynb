```python
# Sütun isimlerini küçük harfli hale getirin
df_cleaned.columns = map(str.lower, df_cleaned.columns)

# Aradaki boşlukları veya "_" karakterlerini kaldırın
df_cleaned.columns = df_cleaned.columns.str.replace(' ', '').str.replace('_', '')
```

```python
df_cleaned['untokenizeddata'] = df_cleaned['booktitle'] + ' ' + df_cleaned['authorname'] + ' ' + df_cleaned['bookdescription'] + ' ' + df_cleaned['genres'] + ' ' + df_cleaned['type']
```

```python
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

def clean_text_and_tokenize(text):
    text = text.lower() # tüm karakterleri küçük harf yap
    # define a translation table with punctuations to remove
    translator = str.maketrans("", "", string.punctuation)
    # use the translation table to remove the punctuations from the text
    text_without_punctuations = text.translate(translator)
    text_without_numbers = re.sub(r'\d+', '', text_without_punctuations)
    text_without_special_chars = re.sub(r'[^a-zA-Z0-9\s]', '', text_without_numbers)
    tokens = nltk.word_tokenize(text_without_special_chars)
    tokens_without_stopwords = [token for token in tokens if not token in stopwords.words('english')]

   
    
    return tokens_without_stopwords
```

```python
df_cleaned['tokenizeddata'] = df_cleaned['untokenizeddata'].apply(lambda x: clean_text_and_tokenize(x))
```

```python
def clean_tokens(tokens):
    tokens = [token.strip() for token in tokens if token.strip() and token.strip() != "''"]

    # remove punctuation from each token, including trailing dots
    tokens = [re.sub(r'[^\w\s]+$', '', token) for token in tokens]

    # convert each token to lowercase
    tokens = [token.lower() for token in tokens]
    tokens = [item for item in tokens if item != '']
    return tokens
```

```python
df_cleaned['tokenizeddata'] = df_cleaned['tokenizeddata'].apply(lambda x: clean_tokens(x))
```
